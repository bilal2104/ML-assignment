```python
import numpy as np  
import matplotlib.pyplot as mtp  
import pandas as pd 
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_squared_error, r2_score

# Load and clean the dataset
train_data_path = 'train.csv'
train_df = pd.read_csv(train_data_path)

# Drop rows with missing values
train_df = train_df.dropna()

# Define feature (X) and target (Y)
x = train_df[['x']].values
y = train_df['y'].values

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)

# Create a Linear Regression model and train it
regressor = LinearRegression()  
regressor.fit(x_train, y_train)  

# Predictions for training and testing data
y_pred = regressor.predict(x_test)  
x_pred = regressor.predict(x_train)  

# Visualize training data and regression line
mtp.scatter(x_train, y_train, color="green")   
mtp.plot(x_train, x_pred, color="red")    
mtp.title("Linear Regression (Training Data)")  
mtp.xlabel("Feature")  
mtp.ylabel("Target")  
mtp.show()   

# Visualize testing data and regression line
mtp.scatter(x_test, y_test, color="blue")   
mtp.plot(x_train, x_pred, color="red")    
mtp.title("Linear Regression (Testing Data)")  
mtp.xlabel("Feature")  
mtp.ylabel("Target")  
mtp.show() 

# Output coefficients and intercept
print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)

# Calculate and output Mean Squared Error and R^2 Score
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Coefficient of Determination (R^2 Score):', r2_score(y_test, y_pred))
```
